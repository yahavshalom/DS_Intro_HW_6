```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
```

## Data

```python
data = ("C:\\Users\\yahav\\MyNotebooks\\LP_destinations.csv")
```

```python
df = pd.read_csv(data,encoding='ISO-8859-1')
df = df.replace(to_replace='Krakï¿', value='krakow', regex=True) ## replace the invalid city
df
```

#### QA 1

```python
lst = list()
for i in df["Description"]:
    lst.append(len(i.split()))
#lst
```

```python
df["num_of_words"] = lst
df 
```

#### QA 2

```python
def remove_punc(my_string):
    my_string = my_string.replace("\n", " ")
    my_string = my_string.replace("'", "")
    my_string = re.sub(r'[^\w\s]',' ',my_string)
    return my_string
```

```python
df["No_punct_description"] = df["Description"].apply(remove_punc)
df
```

#### QA 3

```python
lst3 = list()
for i in df["Description"]:
    num = 0
    split_sentence = i.split()
    for word in split_sentence:
        if word == 'restaurant':
            num += 1
    lst3.append(num)
```

```python
df["has_restaurants"] = lst3
df
```

#### QA 4

```python
lst4 = list()
for i in df["Description"]:
    num = 0
    split_sentence = i.split()
    for word in split_sentence:
        if word == 'museum':
            num += 1
    lst4.append(num)
```

```python
df["has_museums"] = lst4
df
```

#### QA 5

```python
lst5 = list()
for i in df["Description"]:
    num = 0
    split_sentence = i.split()
    for word in split_sentence:
        if word == 'sea' or word == 'beach' or word == 'ocean':
            num += 1
    lst5.append(num)
```

```python
df["has_beaches"] = lst5
df
```
